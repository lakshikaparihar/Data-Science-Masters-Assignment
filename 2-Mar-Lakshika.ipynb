### What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
from matplotlib import pyplot as plt
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of random numbers")
plt.show()
```

### 3. Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

```python
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Line 1")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("Line 2")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("Line 3")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,4)
plt.plot(x,y)
plt.title("Line 4")

plt.subplots_adjust(hspace=0.4)
plt.show()
```

### 4. What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python

plt.subplot(2,1,1)
plt.barh(company ,profit)
plt.title("Horizontal bar graph")

plt.subplot(2,1,2)
plt.bar(company, profit)
plt.title("Verticle bar graph")

plt.subplots_adjust(hspace=0.5)
plt.show()
```
