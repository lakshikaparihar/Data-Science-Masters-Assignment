```python
import seaborn as sns
data = sns.load_dataset('titanic')
```

```python
data.head()
```

### 1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
sns.scatterplot(data=data,x=data['age'],y=data['fare'])
```

```python
sns.displot(data['age'])
```

```python
sns.relplot(x = data['age'],y=data['fare'],data=data,hue='sex')
```

```python
sns.catplot(x = 'class' , y = 'fare' , data=data)
```

```python
sns.jointplot(x='age',y='fare',data=data)
```

### 6. Use the "flights" dataset from seaborn to plot a heatmap.

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
flights_pv = flight.pivot_table(index='month', columns='year', values='passengers')
flights_pv.head()
```

```python
sns.heatmap(data=flights_pv)
```
